```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Chocolate_sales.csv")
df
```

```python
df.info()
```

```python
sales=df.groupby("Sales Person")["Amount"].sum()
plt.figure()
sales.plot(kind="bar")
plt.title("Sales Person Total Amount")
plt.xlabel("Sales Person")
plt.ylabel("Total sales")
plt.show()
```
