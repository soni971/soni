```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("manufacturing_dataset.csv")
df
```

```python
# CONARIANCE

cov_matrix=df.cov()
cov_matrix
```

```python
cor_matric=df.corr()
cor_matric
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.heatmap(cor_matric, annot=True, cmap="plasma")
plt.title("Correlation Matric in Manufacturing")
plt.show()
```
